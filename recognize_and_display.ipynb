```python
import tensorflow as tf
import cv2
import numpy as np
import os
```

```python
model = tf.keras.models.load_model("face_recognition_model.h5")
```

```python
class_names = os.listdir("dataset")
```

```python

student_details = {
    
    "akshat": {"name": "Akshat Gupta", "roll_no": "2K23CSUN01197", "class": "AIML", "section": "A"},
    "aniket": {"name": "Aniket Hans", "roll_no": "2K23CSUN01198", "class": "AIML", "section": "A"},
    "mridul": {"name": "Mridul Kathait", "roll_no": "2K23CSUN01221", "class": "AIML", "section": "A"},
    "tarandeep": {"name": "Tarandeep Singh", "roll_no": "2K23CSUN01233", "class": "AIML", "section": "A"},
}
```

```python

cap = cv2.VideoCapture("http://192.168.126.204:4747/video")

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

print("Camera connected. Press 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        continue

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(50, 50))

    for (x, y, w, h) in faces:
        face_img = frame[y:y+h, x:x+w]
        face_img = cv2.resize(face_img, (224, 224))
        face_img = face_img / 255.0
        face_img = np.expand_dims(face_img, axis=0)

        prediction = model.predict(face_img)[0]
        predicted_index = np.argmax(prediction)
        confidence = prediction[predicted_index]

        if confidence > 0.7:
            folder_name = class_names[predicted_index]
            details = student_details.get(folder_name, {"name": "Unknown", "roll_no": "-", "class": "-"})
            display_text = f"{details['name']} | Roll: {details['roll_no']} | Class: {details['class']}"
        else:
            display_text = "Unknown"

        
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(frame, display_text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    cv2.imshow("Face Recognition Attendance System", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
